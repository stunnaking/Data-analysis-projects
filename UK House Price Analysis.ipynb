```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
#Add Year and Month Column to Dataset
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df.head(2)
```

```python
#Finding out the Average house price in England
df_temp = df[df['area'] == 'england']
df_temp
```

```python
df_temp['average_price'].mean()
```

```python
df_temp.groupby('year')['average_price'].mean()
```

```python
#Average House Price in the entire UK
df.groupby('year')['average_price'].mean()
```

```python
df['area'].unique()
```

```python
#Average house prices according to location
df.groupby('area')['average_price'].mean().sort_values(ascending = False)
```

```python
# Dive into the crime rate in these areas
# Find the MIN, MEAN & MAX of crime in these areas
df.groupby('area')['no_of_crimes'].min().sort_values(ascending = True)
```

```python
df.groupby('area')['no_of_crimes'].max().sort_values(ascending = False)
```

```python
#Years with the most crime recorded
df.groupby('year')['no_of_crimes'].max().sort_values(ascending = False)
```
